```python
# Question 1:Write a python program which searches all the product under a particular product from www.amazon.in.
```

```python
driver=webdriver.Chrome(r"Chromedriver.exe")
```

```python
driver.get("https://www.amazon.in/")
```

```python
Search=driver.find_element(By.CLASS_NAME,"nav-search-submit-text")
Search.click()
```

```python
Product_Name=[]
Price=[]
Expected_Delivery=[]
Product_url=[]
```

```python
start=0
end=3
for i in range(start,end):
    name=driver.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-base a-text-normal"]')
    for i in name[0:10]:
        Product_Name.append(i.text)
   
    Price_tag=driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
    for i in Price_tag[0:10]:
        Price.append(i.text)
    Delivery=driver.find_elements(By.XPATH,'//span[@class="a-color-base a-text-bold"]')
    for i in Delivery[0:10]:
        Expected_Delivery.append(i.text)
    Url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in Url[0:10]:
        Product_url.append(i.get_attribute('href'))
    next_button=driver.find_elements(By.XPATH,"/html/body/div[1]/div[2]/div[1]/div[1]/div/span[1]/div[1]/div[32]/div/div/span/a[3]")
```

```python
len(Product_Name),len(Price),len(Expected_Delivery),len(Product_url)
```

```python
import pandas as pd
df=pd.DataFrame({'BRAND':Product_Name,'PRICE':Price,'Exp_del':Expected_Delivery,'URL':Product_url})
df
```

```python
# Question Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in“London” location
```

```python
driver=webdriver.Chrome(r"Chromedriver.exe")
```

```python
driver.get("https://www.hostelworld.com/")
```

```python
Location=driver.find_element(By.CLASS_NAME,"search-input")
Location.send_keys('London')
```

```python
Hostel_Name=[]
Distance=[]
Ratings=[]
Price=[]
Dorms_Price=[]
```

```python
hostel_tags=driver.find_elements(By.XPATH,'//div[@class="property-name"]')
for i in hostel_tags[0:25]:
    hostel=i.text
    Hostel_Name.append(hostel)
    
Distance_tags=driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
for i in Distance_tags[0:25]:
    dist=i.text
    Distance.append(dist)
    
Ratings_tags=driver.find_elements(By.XPATH,'//div[@class="keyword"]')
for i in Ratings_tags[0:25]:
    rating=i.text
    Ratings.append(rating)
    
price_tags=driver.find_elements(By.XPATH,'//strong[@class="current"]')
for i in price_tags[0:25]:
    price=i.text
    Price.append(price)
    
Dorm_tags=driver.find_elements(By.XPATH,'//strong[@class="current"]')
for i in Dorm_tags[0:25]:
    Dorm=i.text
    Dorms_Price.append(Dorm)
    

    
```

```python
print(len(Hostel_Name),len(Distance),len(Ratings),len(Price),len(Dorms_Price))
```

```python
import pandas as pd
df=pd.DataFrame({'HOSTEL NAME':Hostel_Name,'DISTANCE':Distance,'RATINGS':Ratings,'PRICE':Price,"DORM_PRICE":Dorms_Price})
df
```
